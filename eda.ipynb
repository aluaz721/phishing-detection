```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('phishing_email.csv')
data.head()
```

```python
# Count the number of phishing and legitimate emails
print(data["label"].value_counts())
```

```python
# Look at only emails identified as legitimate (label = 0)
legitimate_emails = data[data['label'] == 0]
legitimate_emails.head()
```

```python
# Look at only emails identified as phishing (label = 1)
phishing_emails = data[data['label'] == 1]
phishing_emails.head()
```

```python
# inspect entire contents of first legitimate email
print(legitimate_emails.iloc[0]['text_combined'])

# inspect entire contents of first phishing email
print(phishing_emails.iloc[0]['text_combined'])
```
